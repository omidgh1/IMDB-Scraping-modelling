## Step 1:
### The first page of the popular movies from 1900 to 2020 is in 'url' list and has 4924 pages next of it. The loop saved next pages in the 'url' list. It takes 30 minutes aproximately on my laptop

```python
url=["https://www.the-numbers.com/movie/budgets/all"]
for i in range (0,62):
    y = str(i*100+1)
    x = url[0]+'/'+y
    url.append (x)
    print (i,':',url[i])
del url[1]
    
```

```python
#Create a handle, page, to handle the contents of the website
dic={}
j=0
for i in url:
    page = requests.get(i)
    #Store the contents of the website under doc
    doc = lh.fromstring(page.content)
    #Parse data that are stored between <tr>..</tr> of HTML
    tr_elements = doc.xpath('//tr')
    dic[j] = tr_elements
    j += 1
dic
```

```python
#Create empty list
col=[]
i=0
#For each row, store each first element (header) and an empty list
for t in dic[0][0]:
    i+=1
    name=t.text_content()
    print ('%d:"%s"'%(i,name))
    col.append((name,[]))
```

```python
for i in dic.values():
    for j in range(1,len(i)):
        T = i[j]
        if len (T) !=6:
            break
            
        z=0
        for t in T.iterchildren():
            data=t.text_content()
            if z>0:
                try:
                    data=int(data)
                except:
                    pass
            col[z][1].append(data)
            z+=1
```

```python
Dict={title:column for (title,column) in col}
df=pd.DataFrame(Dict)
len(df)
```

```python
df.to_csv('budget.csv')
```
