```python
moviePage=[]
count = int(0)
for i in url:
    page = get(i)
    soup = BeautifulSoup(page.content, 'lxml')
    content = soup.find(id='main')
    movieFrame = content.find_all('div',class_= "lister-item mode-advanced")
    for j in range (0,100):
        movieLine = movieFrame[j].find("h3", class_="lister-item-header")
        moviePage.append ("https://www.imdb.com"+(movieLine.find('a').get('href')))
        count += 1     
    print (count)
```

```python
dic = {}
def imdb(x):
    global movieId,title,origtitle,country,location,language,gross,releaseDate
    global runtime,writers,directors,stars,genres,rate,numberVotes,summary
    
    gross, country, language, releaseDate, origtitle, location, runtime = np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan
    
    page = moviePage[x]
    soup = BeautifulSoup(get(page).content, 'lxml')

    movieId = page.split('/')[-2]

#    originalTitle = soup.find("div",  class_="title_wrapper").find('h1').text.replace('\xa0',',').split(',')[0]
    title = soup.find('meta', property="og:title")['content'].split('(')[0][0:-1]

    content = soup.find("div",  class_="article", id='titleDetails')
    for i in range (0,len(content.find_all('div'))):
        x = content.find_all('div')[i].text.split(':')[0]
        if x =='\nCountry':
            country = (content.find_all('div')[i].text.split(':')[1].replace('\n',''))
        elif x =='\nLanguage':
            language = (content.find_all('div')[i].text.split(':')[1].replace('\n',''))
        elif x =='\nRelease Date':
            releaseDate = (content.find_all('div')[i].text.split(':')[1].replace('\n','')).split('(')[0]
        elif x =='\nAlso Known As':
            origtitle = (content.find_all('div')[i].text.split(':')[1].replace('\n','')).split('  ')[0]
        elif x =='\nFilming Locations':
            location = (content.find_all('div')[i].text.split(':')[1].replace('\n','')).split('See more')[0]
        elif x =='\nCumulative Worldwide Gross':
            gross = (content.find_all('div')[i].text.split(':')[1].replace(' ',''))
        elif x =='\nRuntime':
            runtime = (content.find_all('div')[i].text.split(':')[1].replace('\n',''))

    try:
        rate = soup.find("div",  class_="ratingValue").text.replace('\n','').split('/')[0]
    except:
        rate = np.nan

    try:
        numberVotes = soup.find("div",  class_="imdbRating").find('a').text
    except:
        numberVotes = np.nan
    
    directors = np.nan
    writers = np.nan
    stars = np.nan
    
    content=soup.find_all("div", class_="credit_summary_item")
    for i in range (0,len(content)):
        x = content[i].find('h4').text
        if  (x =='Directors:' or x =='Director:'):
            directors=[]
            for j in range(0,(len(content[i].find_all('a')))):
                directors.append (content[i].find_all('a')[j].text)
        elif x == 'Writers:' or x =='Writer:':
            writers=[]
            for j in range(0,(len(content[i].find_all('a')))):
                writers.append (content[i].find_all('a')[j].text)
        elif x == 'Stars:' or x =='Star:':
            stars=[]
            for j in range(0,(len(content[i].find_all('a')))):
                stars.append (content[i].find_all('a')[j].text)

    content=soup.find('div',class_='article',id='titleStoryLine').find_all('div',class_='see-more inline canwrap')
    for i in range (0,len(content)):
        x = content[i].find('h4').text
        genres = []
        if x == 'Genres:':
            for j in range (0,len(content[i].find_all('a'))):
                genres.append (content[i].find_all('a')[j].text)

    summary=soup.find('div',class_='summary_text').text.split('\n')[1].replace('  ','')
    
    dictionary = {'movieId':movieId,'title':title,'originalTitle':origtitle,'country':country,
                 'location':location,'language':language,'gross':gross,'releaseDate':releaseDate,
                 'runtime':runtime,'writers':writers,'directors':directors,'stars':stars,'genres':genres,
                 'rate':rate,'numberVotes':numberVotes,'summary':summary}
    
    return dic.update(dictionary)
```
